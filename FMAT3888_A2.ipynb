```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel("returns.xlsx",
                   sheet_name='Sheet1', index_col='Unnamed: 0')
df.head()
# Assuming 'df' is your DataFrame
# Convert 'DATE' column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# List of asset columns (excluding 'Cash [D]')
assets_columns = [
    'Australian Listed Equity [G]',
    "Int'l Listed Equity (Hedged) [G]",
    "Int'l Listed Equity (Unhedged) [G]",
    'Australian Listed Property [G]',
    "Int'l Listed Property [G]",
    "Int'l Listed Infrastructure [G]",
    'Australian Fixed Income [D]',
    "Int'l Fixed Income (Hedged) [D]",
    "Cash [D]"
]

assets = assets_columns

# Compute log-returns
for col in assets_columns:
    df['log_' + col] = np.log(1 + df[col])

# Define time periods
# Period A: January 2012 to December 2015
start_A = '2012-01-01'
end_A = '2015-12-31'
df_A = df[(df['DATE'] >= start_A) & (df['DATE'] <= end_A)]

# Period B: January 2016 to December 2019
start_B = '2016-01-01'
end_B = '2019-12-31'
df_B = df[(df['DATE'] >= start_B) & (df['DATE'] <= end_B)]

# Columns with log-returns
log_return_columns = ['log_' + col for col in assets_columns]

# Calculate mean vector and covariance matrix for Period A
X_A = df_A[log_return_columns]
a_A = X_A.mean().values  # Mean vector
B_A = X_A.cov().values   # Covariance matrix

# Calculate mean vector and covariance matrix for Period B
X_B = df_B[log_return_columns]
a_B = X_B.mean().values  # Mean vector
B_B = X_B.cov().values   # Covariance matrix

# Display the results
print("Period A Mean Vector (a_A):")
print(pd.Series(np.round(a_A,4), index=assets_columns))

print("\nPeriod A Covariance Matrix (B_A):")
print(np.round(pd.DataFrame(B_A, index=assets_columns, columns=assets_columns).values,4))

print("\nPeriod B Mean Vector (a_B):")
print(pd.Series(np.round(a_B,4), index=assets_columns))

print("\nPeriod B Covariance Matrix (B_B):")
print(np.round(pd.DataFrame(B_B, index=assets_columns, columns=assets_columns).values,4))
```

```python
# Function to compute statistics for a given period
def compute_statistics(a, B, period_name):
    results = {}
    for k in [1, 2]:  # Annual and two-year returns
        n = k * 12  # Number of months
        mu = n * a  # Mean vector scaled by n
        sigma2 = n * np.diag(B)  # Variance vector scaled by n
        cov_matrix = n * B  # Covariance matrix scaled by n
        sigma = np.sqrt(sigma2)  # Standard deviations
        
        # Expected returns
        expected_returns = np.exp(mu + 0.5 * sigma2) - 1
        
        # Variances
        variances = np.exp(2 * mu + 2 * sigma2) - np.exp(2 * mu + sigma2)
        
        # Prepare to compute covariances and correlations
        expected_returns_matrix = np.outer(np.exp(mu + 0.5 * sigma2), np.exp(mu + 0.5 * sigma2))
        covariances = np.exp(mu.reshape(-1, 1) + mu + 0.5 * (sigma2.reshape(-1, 1) + sigma2)) * (np.exp(cov_matrix) - 1)
        # Alternatively, compute covariances using the formula directly
        # covariances = expected_returns_matrix * (np.exp(cov_matrix) - 1)
        
        # Adjust variances on the diagonal
        np.fill_diagonal(covariances, variances)
        
        # Correlations
        std_devs = np.sqrt(np.diag(covariances))
        correlation_matrix = covariances / np.outer(std_devs, std_devs)
        
        # Store results
        results[f'Period {period_name}, Return {k} Year(s)'] = {
            'Expected Returns': pd.Series(expected_returns, index=assets),
            'Covariance Matrix': pd.DataFrame(covariances, index=assets, columns=assets),
            'Correlation Matrix': pd.DataFrame(correlation_matrix, index=assets, columns=assets)
        }
    return results

# Compute statistics for Period A
stats_A = compute_statistics(a_A, B_A, 'A')

# Compute statistics for Period B
stats_B = compute_statistics(a_B, B_B, 'B')

# Display results
for key, value in stats_A.items():
    print(f"\n{key}:")
    print("\nExpected Returns:")
    print(value['Expected Returns'])
    print("\nCovariance Matrix:")
    print(np.round(value['Covariance Matrix'].values,4))
    print("\nCorrelation Matrix:")
    print(np.round(value['Correlation Matrix'].values,4))

# Similarly for Period B
for key, value in stats_B.items():
    print(f"\n{key}:")
    print("\nExpected Returns:")
    print(value['Expected Returns'])
    print("\nCovariance Matrix:")
    print(np.round(value['Covariance Matrix'].values,4))
    print("\nCorrelation Matrix:")
    print(np.round(value['Correlation Matrix'].values,4))
```

```python
from scipy.optimize import minimize

# Extract data for Period A
mu_A = stats_A['Period A, Return 2 Year(s)']['Expected Returns']
cov_A = stats_A['Period A, Return 2 Year(s)']['Covariance Matrix']

# Extract data for Period B
mu_B = stats_B['Period B, Return 2 Year(s)']['Expected Returns']
cov_B = stats_B['Period B, Return 2 Year(s)']['Covariance Matrix']


# Define the objective function
def objective_function(w, mu, cov):
    return -w.T @ mu + 0.5 * w.T @ cov @ w

# Equality constraint: sum of weights equals 1
def constraint_sum_weights(w):
    return np.sum(w) - 1

# Bounds for weights: w_i >= 0
bounds = [(0, 1) for _ in range(len(mu_A))]

# Constraints dictionary
constraints = [
        {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},  # Sum of weights =1
        {'type': 'ineq', 'fun': lambda w: w[0] - 0.15},  # Australian Listed Equity [G] >=15%
        {'type': 'ineq', 'fun': lambda w: w[1] - 0.15},  # Int'l Listed Equity (Hedged) [G] >=15%
        {'type': 'ineq', 'fun': lambda w: w[6] - 0.05},  # Australian Fixed Income [D] >=5%
        {'type': 'ineq', 'fun': lambda w: w[8] - 0.05},  # Cash [D] >=5%
        # Optional: Growth Assets between 65% and 75%
        {'type': 'ineq', 'fun': lambda w: np.sum(w[0:6]) - 0.65},  # >=65%
        {'type': 'ineq', 'fun': lambda w: 0.75 - np.sum(w[0:6])}   # <=75%
    ]

# Initial guess (equal weights)
w0 = np.ones(len(mu_A)) / len(mu_A)

def portfolio_statistics(w, mu, cov):
    expected_return = w.T @ mu
    variance = w.T @ cov @ w
    volatility = np.sqrt(variance)
    return expected_return, volatility

# Optimize for Period A
result_A = minimize(objective_function, w0, args=(mu_A, cov_A), method='SLSQP',
                    bounds=bounds, constraints=constraints)

# Display results for Period A
if result_A.success:
    w_opt_A = result_A.x
    w_opt_A /= np.sum(w_opt_A)  # Normalize weights
    expected_return_A, volatility_A = portfolio_statistics(w_opt_A, mu_A, cov_A)
    weights_A = pd.DataFrame({'Asset': assets, 'Weight': w_opt_A})
    print("Optimal Weights for Period A (using scipy.optimize.minimize):")
    print(np.round(weights_A, 4))
    print(f"Objective (Period A): {result_A.fun:.4f}")
    print(f"Expected Return (Period A): {expected_return_A:.4f}")
    print(f"Volatility (Period A): {volatility_A:.4f}")
else:
    print("Optimization for Period A did not converge:")
    print(result_A.message)
    
result_B = minimize(objective_function, w0, args=(mu_B, cov_B), method='SLSQP',
                    bounds=bounds, constraints=constraints)

# Display results for Period B
if result_B.success:
    w_opt_B = result_B.x
    w_opt_B /= np.sum(w_opt_B)  # Normalize weights
    expected_return_B, volatility_B = portfolio_statistics(w_opt_B, mu_B, cov_B)
    weights_B = pd.DataFrame({'Asset': assets, 'Weight': w_opt_B})
    print("\nOptimal Weights for Period B (using scipy.optimize.minimize):")
    print(np.round(weights_B, 4))
    print(f"Objective (Period B): {result_B.fun:.4f}")    
    print(f"Expected Return (Period B): {expected_return_B:.4f}")
    print(f"Volatility (Period B): {volatility_B:.4f}" )
```

```python
import numpy as np
from scipy.optimize import minimize
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Mean and covariance for asset returns
mu = stats_B['Period B, Return 1 Year(s)']['Expected Returns']
cov_matrix = stats_B['Period B, Return 1 Year(s)']['Covariance Matrix']
cov_matrix.iloc[:, 8] = 0
cov_matrix.iloc[8, :] = 0

# Parameters
n_assets = 9
N = 10000  # Number of simulated scenarios
gamma = 1  # Risk aversion coefficient

# Simulate V1 and V2 as i.i.d.
V1 = multivariate_normal.rvs(mean=mu, cov=cov_matrix, size=N)
V2 = multivariate_normal.rvs(mean=mu, cov=cov_matrix, size=N)

def utility(x, gamma=1):
    return -np.exp(-gamma * x)

def portfolio_performance(weights, returns):
    """
    Calculate expected return and variance of a portfolio.
    """
    port_return = np.dot(weights, returns.mean(axis=0))
    port_variance = np.dot(weights, np.dot(returns.cov(), weights))
    return port_return, port_variance

def efficient_frontier(returns, num_portfolios=100):
    """
    Compute the efficient frontier.
    """
    results = np.zeros((3, num_portfolios))
    weights_record = []
    target_returns = np.linspace(returns.mean().min(), returns.mean().max(), num_portfolios)
    
    for i, target in enumerate(target_returns):
        # Objective: minimize variance
        def objective(weights):
            return np.dot(weights, np.dot(returns.cov(), weights))
        
        # Constraints: sum(weights) = 1 and expected return = target
        constraints = [
            {'type': 'eq', 'fun': lambda w: np.sum(w) - 1},
            {'type': 'eq', 'fun': lambda w: np.dot(w, returns.mean(axis=0)) - target},
            {'type': 'ineq', 'fun': lambda w: w[0] - 0.15},  # Australian Listed Equity >=15%
            {'type': 'ineq', 'fun': lambda w: w[1] - 0.15},  # Int'l Listed Equity >=15%
            {'type': 'ineq', 'fun': lambda w: w[6] - 0.05},  # Australian Fixed Income >=5%
            {'type': 'ineq', 'fun': lambda w: w[8] - 0.05},  # Cash >=5%
            {'type': 'ineq', 'fun': lambda w: np.sum(w[0:6]) - 0.65},  # Growth assets >=65%
            {'type': 'ineq', 'fun': lambda w: 0.75 - np.sum(w[0:6])}  # Growth assets <=75%
        ]
        
        # Bounds: no short-selling
        bounds = tuple((0, 1) for _ in range(n_assets))
        
        # Initial guess: equally distributed
        initial_guess = np.full(n_assets, 1/n_assets)
        
        # Optimize
        result = minimize(objective, initial_guess, method='SLSQP', bounds=bounds, constraints=constraints)
        
        if result.success:
            weights = result.x
            port_return, port_variance = portfolio_performance(weights, returns)
            results[0, i] = port_return
            results[1, i] = port_variance
            results[2, i] = port_return / np.sqrt(port_variance)  # Sharpe ratio
            weights_record.append(weights)
    
    return results, weights_record

# Define the utility function
def expected_utility(w, u, V1, V2, gamma=1):
    G = (1 + np.dot(V1, w)) * (1 + np.dot(V2, u)) - 1
    U = utility(G, gamma)
    return np.mean(U), G  # Return both utility and G values

# Function to optimize u given w
def optimize_u_given_w(w, V2, gamma=1):
    """
    Optimize second-period portfolio weights u given first-period weights w.
    """
    def objective(u):
        G = (1 + np.dot(V1, w)) * (1 + np.dot(V2, u)) - 1
        U = utility(G, gamma)
        return -np.mean(U)
    
    # Constraints: sum(u) = 1
    constraints = [
        {'type': 'eq', 'fun': lambda u: np.sum(u) - 1},
        {'type': 'ineq', 'fun': lambda u: u[0] - 0.15},
        {'type': 'ineq', 'fun': lambda u: u[1] - 0.15},
        {'type': 'ineq', 'fun': lambda u: u[6] - 0.05},
        {'type': 'ineq', 'fun': lambda u: u[8] - 0.05},
        {'type': 'ineq', 'fun': lambda u: np.sum(u[0:6]) - 0.65},
        {'type': 'ineq', 'fun': lambda u: 0.75 - np.sum(u[0:6])}
    ]
    
    # Bounds: no short-selling
    bounds = tuple((0, 1) for _ in range(n_assets))
    initial_guess = np.full(n_assets, 1/n_assets)
    
    result = minimize(objective, initial_guess, method='SLSQP', bounds=bounds, constraints=constraints)
    
    return result.x if result.success else np.full(n_assets, 1/n_assets)

# Function to find the optimal w by iterating over the efficient frontier
def find_optimal_w(frontier_results, frontier_weights, V1, V2, gamma=1):
    max_utility = -np.inf
    optimal_w = None
    optimal_u = None
    G_values = None
    
    for i, w in enumerate(frontier_weights):
        u = optimize_u_given_w(w, V2, gamma)
        eu, G = expected_utility(w, u, V1, V2, gamma)
        
        if eu > max_utility:
            max_utility = eu
            optimal_w = w
            optimal_u = u
            G_values = G  # Store G values
    
    return optimal_w, optimal_u, max_utility, G_values

# Create a DataFrame for returns to use with portfolio functions
returns_df = pd.DataFrame(V1, columns=[f'Asset_{i+1}' for i in range(n_assets)])

# Compute efficient frontier for the first period
frontier_results, frontier_weights = efficient_frontier(returns_df)

# Find the optimal w and u
optimal_w, optimal_u, max_utility, G_values = find_optimal_w(frontier_results, frontier_weights, V1, V2, gamma)

# Output optimal weights and utility
print("Optimized first-period weights (w):")
print(np.round(optimal_w, 4))
print("\nOptimized second-period weights (u):")
print(np.round(optimal_u, 4))
print(f"\nMaximum Expected Utility: {max_utility}")

# Calculate and print expected return and standard deviation for the total return G
G_mean = np.mean(G_values)
G_std = np.std(G_values)

print(f"\nTotal Return G over two years:")
print(f"Mean: {G_mean:.4f}")
print(f"Standard Deviation: {G_std:.4f}")

# Plot histogram of G values (total returns)
plt.hist(G_values, bins=50, edgecolor='k')
plt.title('Histogram of Total Returns over Two Years')
plt.xlabel('Total Return G')
plt.ylabel('Frequency')
plt.show()
```
